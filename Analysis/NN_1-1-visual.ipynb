```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score, ConfusionMatrixDisplay
from typing import Dict, List
from matplotlib import pyplot as plt

df = pd.read_csv(filepath_or_buffer="NN_confusion_file.csv", names=["Feature", "Accuracy", "f1", "TN", "FP", "FN", "TP"], sep=", ", header=0)
df
```

# Numerical features

```python
numerical_columns: List[str] = [f"f_{i}" for i in range(42, 80)]
```

```python
fig, axs = plt.subplots(8,5, figsize=(20,20))
for i in range(len(numerical_columns)):
    frow = df[df["Feature"] == numerical_columns[i]]
    cf = np.matrix([
    [frow["TN"].item(), frow["FP"].item()],
    [frow["FN"].item(), frow["TP"].item()]
    ])
    ax: plt.Axes = axs[i // 5, i % 5]
    ax.set_title(f"{numerical_columns[i]}")
    ConfusionMatrixDisplay(confusion_matrix=cf).plot(ax=ax)
plt.show()
```

# Categorical features
